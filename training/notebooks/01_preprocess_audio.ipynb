# Notebook 01 – Prétraitement audio

```python
# Prétraitement des fichiers audio en Mel-spectrogrammes
import librosa
import numpy as np
import os

def load_and_preprocess(file_path, sr=16000, duration=3):
    y, _ = librosa.load(file_path, sr=sr, duration=duration)
    if len(y) < sr*duration:
        y = np.pad(y, (0, sr*duration - len(y)))
    else:
        y = y[:sr*duration]
    mel = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128, n_fft=2048, hop_length=512)
    mel_db = librosa.power_to_db(mel, ref=np.max)
    mel_db = mel_db[:128, :124]  # taille fixe
    return mel_db[..., np.newaxis]

# Appliquer à tous les fichiers
data_dir = "../data/audio"
for cls in ['normal', 'cry', 'fight']:
    path = os.path.join(data_dir, cls)
    if not os.path.exists(path): continue
    for f in os.listdir(path):
        if f.endswith('.wav'):
            mel = load_and_preprocess(os.path.join(path, f))
            np.save(os.path.join(path, f.replace('.wav', '.npy')), mel)
```
