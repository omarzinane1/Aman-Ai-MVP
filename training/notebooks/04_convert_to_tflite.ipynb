# Notebook 04 – Conversion vers TensorFlow Lite

```python
import tensorflow as tf
import os

os.makedirs('../models', exist_ok=True)

# Conversion audio avec quantification INT8
if os.path.exists('../models/audio_model.h5'):
    model = tf.keras.models.load_model('../models/audio_model.h5')
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    converter.optimizations = [tf.lite.Optimize.DEFAULT]
    converter.representative_dataset = lambda: [tf.random.uniform((1,128,124,1))]
    converter.target_spec.supported_ops = [tf.lite.OpsSet.TFLITE_BUILTINS_INT8]
    converter.inference_input_type = tf.int8
    converter.inference_output_type = tf.int8
    tflite_model = converter.convert()
    with open('../models/audio_model.tflite', 'wb') as f:
        f.write(tflite_model)

# Conversion mouvement (float32)
if os.path.exists('../models/motion_model.h5'):
    model = tf.keras.models.load_model('../models/motion_model.h5')
    converter = tf.lite.TFLiteConverter.from_keras_model(model)
    tflite_model = converter.convert()
    with open('../models/motion_model.tflite', 'wb') as f:
        f.write(tflite_model)
```
